```python
import spacy
from spacy import displacy
from IPython.display import HTML, display

# Load the English language model
nlp = spacy.load("en_core_web_sm")

text = "I saw the man with a telescope."

doc = nlp(text)

# Print the dependency parse information for each token
for token in doc:
    print(f"{token.text:<12} {token.dep_:<12} {token.head.text}")

html = displacy.render(doc, style="dep", jupyter=False)

# Display the dependency parse visualization in Jupyter Notebook
display(HTML(html))
```

```python

# 1. What is the ROOT? - saw
# 2. What does "with a telescope" modify? - saw
# 3. Does the sentence have more than one meaning? - Yes Either I had the telescope, or the man did.
# 4. Did spaCy choose the interpretation you expected?
# 5. Rewrite the sentence to remove ambiguity.- (a) → "Using a telescope, I saw the man.", (b) → "I saw the man who had a telescope."
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """
Microsoft announced a new AI research initiative in Berlin on 15 June 2026.
The project involves collaboration with Google, SAP and SRH University Heidelberg.
Satya Nadella will present the initiative at a conference in Germany.
"""

doc = nlp(text)

for ent in doc.ents:
    print(
        ent.text,
        ent.label_,
        spacy.explain(ent.label_)
    )
```

```python
#1. Which entities were detected? - Microsoft, Berlin, 15 June 2026, Google, SAP, SRH University Heidelberg, Satya Nadella, Germany
#2. Which labels were assigned? - ORG, GPE, DATE, PERSON
#3. Which entity types occur most often? - ORG — 4 times
#4. Were any entities missed? Yes — "AI"
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "AI is transforming healthcare",
    "Machine learning improves healthcare",
    "Football is a popular sport"
]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())

similarity_matrix = cosine_similarity(tfidf_matrix)

print("\nCosine Similarity Matrix:")
print(similarity_matrix)

print("Doc1 vs Doc2:", round(similarity_matrix[0][1], 3))
print("Doc1 vs Doc3:", round(similarity_matrix[0][2], 3))
print("Doc2 vs Doc3:", round(similarity_matrix[1][2], 3))
```

```python
# 1. Which pair has the highest similarity? - Doc1 vs Doc2 — score 0.172 
# 2. Is this result reasonable? - Yes — Doc1 & Doc2 both talk about healthcare. Doc2 vs Doc3 = 0.0
# 3. Why does TF-IDF treat some words as more important? - Rare words score higher "transforming" appears in only 1 doc → high score. "healthcare" appears in 2 docs → lower score. Common = less informative.
# 4. Does cosine understand meaning? - No pure vector math, no semantics Just compares number overlap
# 5. Would it know that "AI" and "machine learning" are related? - No different strings = zero overlap = 0.0 similarity
```
